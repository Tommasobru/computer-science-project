### 7. Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)
1. Converts the ‘tpep_pickup_datetime’ column to datetime format, if it is not already.
   This is essential to be able to do time operations.
2. Round each pickup time down to the nearest 30 minute block.
3. Calculate the upper bound of the interval by adding 30 minutes to the start_time.

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])


start_time = df['tpep_pickup_datetime'].dt.floor('30min')

end_time = start_time + pd.Timedelta(minutes=30)

df['pickup_interval'] = start_time.dt.strftime('%H:%M') + ' - ' + end_time.dt.strftime('%H:%M')

df[['tpep_pickup_datetime','pickup_interval']]
```

1. avg_fare: is a DataFrame that contains the average of the fare_amount for each payment_type and pickup_interval on the common dataset (i.e. the subset of trips to the 5 most common destinations for each pickup location).

### 17. Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.
1. Converts the tpep_pickup_datetime and tpep_dropoff_datetime columns to datetime64 format to allow time operations.
2. Sort trips by VendorID and in ascending chronological order.
3. Extracting columns as NumPy arrays: it is used to speed up array operations rather than working on DataFrame directly.
4. Initializing the chain array: each row will have a value in the chain column that identifies the chain it belongs to. Everything is initially 0.
5. Chain construction:
   - If the trip has not yet been assigned to a chain, create a new one.
   - Look for a next trip that: has the same VendorID, pickup occurs within 2 minutes of the current dropoff and the pickup location is the same as the current dropoff location.
   - bisect_left(a, x) returns the leftmost index where you can insert x into the sorted list a without changing the order.
   - we efficiently find rows starting between dropoff_time and 2 minutes later.
   - Assign the same chain_id to the next trip if it meets the conditions.

```python
import numpy as np
import pandas as pd
from bisect import bisect_left, bisect_right

# Convert datetime columns to proper format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Sort by VendorID and pickup time
df = df.sort_values(by=["VendorID", "tpep_pickup_datetime"]).reset_index(drop=True)

# Extract columns as NumPy arrays
pickup_times = df['tpep_pickup_datetime'].values
dropoff_times = df['tpep_dropoff_datetime'].values
vendor_ids = df['VendorID'].values
pulocation_ids = df['PULocationID'].values
dolocation_ids = df['DOLocationID'].values

# Initialize chain array
chains = np.zeros(len(df), dtype=np.int32)
current_chain_id = 1

# Loop through all rows
for i in range(len(df)):
    if chains[i] == 0:
        chains[i] = current_chain_id
        current_chain_id += 1

    # Current row values
    cur_vendor = vendor_ids[i]
    cur_dropoff_time = dropoff_times[i]
    cur_dolocation = dolocation_ids[i]
    cur_chain = chains[i]

    # Binary search to find window of candidate rows
    start_idx = bisect_left(pickup_times, cur_dropoff_time, lo=i+1)
    end_idx = bisect_right(pickup_times, cur_dropoff_time + np.timedelta64(120, 's'), lo=start_idx)

    # Iterate only through the candidate window
    for j in range(start_idx, end_idx):
        # Break early if VendorID changes (dataset is sorted)
        if vendor_ids[j] != cur_vendor:
            break

        if pulocation_ids[j] == cur_dolocation and chains[j] == 0:
            chains[j] = cur_chain  # assign same chain
            break  # only the first match within 2 minutes

# Assign back to dataframe
df['chain'] = chains
```

```python
# Conta i viaggi per ogni catena
chain_counts = df['chain'].value_counts()

# Prendi solo le chain con almeno 3 viaggi
valid_chains = chain_counts[chain_counts > 2].index

# Filtra il DataFrame per mostrare solo quelle righe
df_filtered = df[df['chain'].isin(valid_chains)]

df_filtered = df_filtered.sort_values(by=['chain', 'tpep_pickup_datetime']).reset_index(drop=True)
df_filtered[['VendorID', 'tpep_pickup_datetime',	'tpep_dropoff_datetime', 'PULocationID', 'DOLocationID', 'chain']]
```
